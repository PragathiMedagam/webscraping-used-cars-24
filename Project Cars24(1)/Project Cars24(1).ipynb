## Data Cleaning   

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import plotly.express as px
import warnings 
warnings.filterwarnings("ignore")
```

```python
df1=pd.read_csv("Used-Car-Details1.csv")
df1
```

```python
df1.info()
```

### Checking null values

```python
df1.isnull().sum()
```

#### Creating new columns

```python
s=df1["Description"].str.split(" ",n=-1,expand=True)
```

```python
df1["Year"]=s[0]
df1["Brand"]=s[1]
df1["Model"]=s[2]
```

```python
df1["Brand"].value_counts()
```

```python
df1["Brand"]= df1["Brand"].str.title()
df1
```

```python
df1["Brand"].value_counts()
```

```python
df1["Model"].value_counts()
```

#### Split EMI value

```python
df1["EMI"].replace(to_replace=["₹",",","/month"],value="",regex=True,inplace=True)
df1
```

#### Split original price

```python
df1["Original_price"].replace(to_replace=["₹",","],value="",regex=True,inplace=True)
df1
```

#### Split specifications

#### Separate Kms

```python
regex = r'([,\d]+\skm)'
df1['km'] = df1.Specification.apply(lambda x : re.findall(regex,x) if x!="NaN" else x)
df1.km = df1.km.apply(lambda x: "".join(x))
df1['km'].replace(to_replace=["km",','],value="",inplace=True,regex=True)
```

```python
df1
```

#### Separate Owners

```python
regex=r"(?:1st|2nd)"
df1["Owner"]=df1.Specification.apply(lambda x : re.findall(regex,x) if x!="NaN" else x)
```

```python
df1["Owner"]=df1["Owner"].apply(lambda x : "".join(x))
df1["Owner"].replace(to_replace=[",","km"],value="",inplace=True,regex=True)
```

```python
df1
```

```python
df1.Owner.replace('',np.nan,inplace=True,regex = True)
```

```python
df1
```

#### Fuel Type

```python
regex = r'(?:Petrol|Diesel)'
df1['fuel_Type'] = df1["Specification"].apply(lambda x : re.findall(regex,x) if x!='NaN' else x)
df1["fuel_Type"] = df1["fuel_Type"].apply(lambda x: "".join(x))
df1
```

#### Car Type

```python
regex = r'(?:Manual|Automatic)'
df1['Type'] = df1["Car_Type"].apply(lambda x : re.findall(regex,x) if x!='NaN' else x)
df1["Type"] = df1["Type"].apply(lambda x: "".join(x))
df1
```

#### Engine Type

```python
df1["Car_Type"].replace(to_replace=["Manual","Automatic"],value="",inplace=True,regex=True)
df1
```

### drop Description,Specification

```python
df1.drop(["Description","Specification","City"],axis=1,inplace=True)
```
