```python
citycode = [2,134,5,3686,4709,2738,5732,777,2423,2987,769,6,290,8184,1605]
cities = ['New Delhi','Noida','Gurgon','Hyderabad','Banglore','Mumbai','Chennai','Kolkata','Pune','Bhopal','Chandigarh','Faridabad','Lucknow','patna','Surat']
brands = ['TOYOTA','MARUTI','VOLKSWAGEN','HONDA','HYUNDAI','KIA','FORD','MAHINDRA','RENAULT','SKODA','BMW','VOLVO','BAJAJ']
description=[]
car_type=[]
emi=[]
original_price=[]
specification=[]
city = []
for b in brands:
    for c in citycode:
        url = 'https://www.cars24.com/buy-used-car?sort=P&search={}&storeCityId={}'.format(b,c)
        
        page = requests.get(url)
        soup = BeautifulSoup(page.content)        
    # DEscription
        for i in soup.find_all("h2",class_="_3FpCg"):
            description.append(i.text)
            city.append(c)
        #Car_type
        for i in soup.find_all("p",class_="cvakB"):
            car_type.append(i.text)   
        #EMI\
        for i in soup.find_all("div",class_="_2HFRN"):
            emi.append(i.text)
        #original_price
        for i in soup.find_all("div",class_="_7udZZ"):
            original_price.append(i.text)
        #specification
        for i in soup.find_all("ul",class_="bVR0c"):
            specification.append(i.text)    
```

```python
print(len(description))
print(len(car_type))
print(len(specification))
print(len(car_type))
print(len(original_price))
print(len(emi))
```

```python
df1 = pd.DataFrame({'Description': description,
                    'Car_Type'  :  car_type,
                    'EMI':   emi,
                    'Original_price':original_price,
                    'Specification':specification,
                    'City' :    city})
```
